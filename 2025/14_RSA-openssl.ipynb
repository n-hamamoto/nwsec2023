# RSA暗号のデモ

opensslコマンドを利用して，公開鍵や秘密鍵を利用してRSA暗号化を行ってみる。使われている素数も確認する。

## 秘密鍵生成

512の部分が鍵長(bit)。最近だと2048bit以上が推奨されている

```python
#! openssl genrsa 2048 > private.key
! openssl genrsa 512 > private.key
```

```python
# 秘密鍵の確認
! cat private.key
```

```python
# 見やすく表示する
!openssl rsa -text < private.key
```

- modulus: 二つの素数の積($n=pq$)。公開しても良い値。

- publicExponent: $(p-1)(q-1)$と互いな素な正整数$e$。公開しても良い値

- privateExponent: $ed \equiv 1 \pmod{(p-1)(q-1)}$となる正整数$d$

- prime1: 素数1 $p$

- prime2: 素数2 $q$
---
- exponent1: $d$ mod $(p-1)$
- exponent2: $d$ mod $(q-1)$
- coefficient: $cq\equiv 1$ mod $p$なる正整数$c$ 

### p（素数1: prime1）

```python
# 証明書の16進数をコピー。改行は詰めること。
# 入力例 a='00:d9:de:d8:a5:bf:0e:6b:98:0d:ca:39:78:93:1d:d4:2e:ed:5a:8b:d0:d4:fd:2a:4f:f9:80:bd:d3:a4:4c:50:c7'

a='00:fa:1f:ef:2c:07:a8:4a:53:c7:4d:e2:ee:a1:62:ae:6a:61:cd:63:3e:69:5a:17:e7:8b:77:de:ff:63:fa:0c:4f'
p=int(a.replace(':', ''), 16)
print(p)
```

### q（素数2: prime2）

```python
# 証明書の16進数をコピー。改行は詰めること。
# 入力例 a='00:d9:de:d8:a5:bf:0e:6b:98:0d:ca:39:78:93:1d:d4:2e:ed:5a:8b:d0:d4:fd:2a:4f:f9:80:bd:d3:a4:4c:50:c7'

a='00:c8:e7:65:0c:59:71:0c:46:08:12:eb:04:c0:8e:29:53:81:06:d0:0d:49:55:d8:e7:17:b5:6a:fe:32:12:9f:19'
q=int(a.replace(':', ''), 16)
print(q)
```

## 公開鍵の作成

```python
#公開鍵作成
! openssl rsa -pubout < private.key > public.key
```

```python
! openssl rsa -text -pubin < public.key
```

秘密鍵にも出てきましたが：

- modulus: 二つの素数の積(𝑛=𝑝𝑞)。公開しても良い値。

- Exponent: 𝑒。公開しても良い値

## 公開鍵で暗号と復号

### 公開鍵で暗号化

```python
#暗号化
! echo "HELLO" | openssl rsautl -encrypt -pubin -inkey public.key > message
```

```python
# 暗号化されたファイル（message）を16進数で表示
! od -x message
```

### 対応する秘密鍵で復号

```python
#復号
! cat message | openssl rsautl -decrypt -inkey private.key
```
