## RRの指定
RR（リソースレコード）を指定して検索できます。

```python
! dig kakaku.com
```

MXレコードを検索

```python
! dig MX kakaku.com
```

全てのRR（リソースレコード）を表示

```python
! dig ANY kakaku.com
```
