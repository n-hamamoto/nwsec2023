## ルート情報

デフォルトゲートウエイが分かります。確認してみましょう。

```python
! ip route
```

8.8.8.8へ通信するための送信ルートを表示します。

8.8.8.8は自身のネットワークの外なので，デフォルトルート（デフォルトゲートウエイ）に送信されると出てきます。

```python
! ip route get 8.8.8.8
```

10.100.1.30宛ての場合は，自分のネットワーク（10.1.0.0/20）なので，直接宛先に送信すると出てきます。

```python
! ip route get 10.100.1.30
```

他の場合も試してみましょう。

```python
! ip route get 172.18.0.4
```

```python
! ip route get 10.2.0.0
```
