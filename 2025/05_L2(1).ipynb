`link/ether xx:xx:xx:xx:xx:xx`の部分がMACアドレスです。

３つあると思いますが、lo:の`00:00:00:00:00:00`はループバックインタフェースと言って、自分自身としか通信できないインタフェースなので無視してください。

- MACアドレスは，グローバル/ローカルのどちらでしょうか？
- MACアドレスは，ユニキャストでしょうか？

```python
# 16進数を2進数に変換する例（16進数の1dを２進数にする）
bin(0x9c)

#出力結果の頭にある`0b`は2進数の意味。
```

例えばMACアドレスの最初の8bitが`1d`だとすると、上のセルで2進数に変換すると`10011100`なので，
- I/Gビット（一番右）が0（ユニキャスト）
- U/Lビット（右から2番目）が0（グローバル）
