# パケット観察

パケットを眺めてみて、講義で話した内容とあっているか確認してみましょう。

パケットの観察には`tcpdump`を使います。`tcpdump`はネットワークインタフェースに流れてくるパケットの内容をそのまま表示するツールです。

## ネットワークインターフェースの確認

````python
! ip a
````

## tcpdumpの使い方

### 実行例1

```
tcpdump -nn -c 10 -x -v -e -i eth0
```
ネットワークインターフェースeth0を通過するパケットを10個キャプチャして終了します。
```
コマンドオプションの意味：
 -nn     ip/portの名前変換しない
 -c 10   10パケットをキャプチャしたら終了
 -e      Ethernetヘッダ情報も出力
 -x      パケット内容を16進数でダンプ（-Xだと16進数とはASCIIでダンプ）
 -i eth0 
```

````python
! sudo tcpdump -nn -c 10 -x -v -e -i eth0
````

### 実行例2

```
tcpdump -nn -c 10 -x -v -e -i eth1 
```
ネットワークインターフェースeth1を通過するパケットを10個キャプチャして終了します。

### 実行例3

```

```
ネットワークインターフェースeth1を通過する**arpフレーム**を10個キャプチャして終了します。
