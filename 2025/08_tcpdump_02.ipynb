パケットの観察には`tcpdump`を使います。`tcpdump`はネットワークインタフェースに流れてくるパケットの内容をそのまま表示するツールです。

## tcpdumpの使い方

```
tcpdump -nn -c 10 -x -vvv -e -i eth0
```
ネットワークインターフェースeth0を通過するパケットを10個キャプチャして終了します。
```
コマンドオプションの意味：
 -nn     ip/portの名前変換しない
 -c 10   10パケットをキャプチャしたら終了
 -e      Ethernetヘッダ情報も出力
 -x      パケット内容を16進数でダンプ（-Xだと16進数とはASCIIでダンプ）
 -vvv    詳細を表示
 -S      Sequence Numberをそのまま表示
 -s      キャプチャするサイズ(byte)を指定(-s0だとすべてキャプチャ, -s128だと128byte）
 -i eth0 
```
